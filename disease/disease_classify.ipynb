# data prep

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import math
```

```python
dataframe = pd.read_csv("data/train.csv")

# delete broken data
dataframe = dataframe.drop("Unnamed: 133", axis=1) # problem with the dataset
dataframe = dataframe.dropna(how='any',axis=0)
```

```python
# replace categorical data with discrete whole numbers
categorical_column_keys = ["prognosis"]
categorical_unique_values = {}

for i in categorical_column_keys:
    unique = dataframe[i].unique()
    categorical_unique_values[i] = unique
    dataframe[i].replace(unique, np.arange(0, len(dataframe[i].unique()), 1, dtype=int), inplace=True)
```

```python
OUTPUT_KEY = "prognosis"

X_dataframe = dataframe.drop(OUTPUT_KEY, axis=1)
X = X_dataframe.values
y = dataframe[OUTPUT_KEY].values
```

```python
dataframe
```

# train

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

# test

```python
test_dataframe = pd.read_csv("data/test.csv")
test_dataframe = test_dataframe.dropna(how='any',axis=0)

# replace categorical data with discrete whole numbers
categorical_unique_values = {}

for i in categorical_column_keys:
    unique = test_dataframe[i].unique()
    categorical_unique_values[i] = unique
    test_dataframe[i].replace(unique, np.arange(0, len(test_dataframe[i].unique()), 1, dtype=int), inplace=True)

OUTPUT_KEY = "prognosis"

X_test_dataframe = test_dataframe.drop(OUTPUT_KEY, axis=1)
X_test = X_test_dataframe.values
y_test = test_dataframe[OUTPUT_KEY].values
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
x = np.full(len(X_dataframe.keys()), 0)
print(X_dataframe.keys()[model.predict([x])])
```
